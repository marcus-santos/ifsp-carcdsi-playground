```python
import pandas
```

```python
# Cria um Dataframe

datas = {'nome': ['Argentina', 'Brasil', 'França', 'Itália', 'Reino Unido'],
          'continente': ['América', 'América', 'Europa','Europa', 'Europa'],
          'extensao': [2780,8511,644,301,244],
          'corVerde': [0,1,0,1,0]
         }

codes = ['AR', 'BR', 'FR', 'IT', 'UK']

countries = pandas.DataFrame(datas, index=codes)

print(countries)

countries
```

```python
countries.info()
```

```python
# Propriedades básicas dos DataFrames
numLines = countries.shape[0]
numCols = countries.shape[1]
position = countries.index
columns = countries.columns
countriesType = type(countries)
countriesDType = countries.dtypes
countriesIdxType = countries.index.dtype

print('Número de linhas: ', numLines)
print('Número de colunas: ', numCols)
print('Rótulo das linhas: ', position)
print('Rótulos das colunas: ', columns)
print('Tipo (type): ', countriesType)
print('dtype das colunas: ', countriesDType)
print('dtype dos rótulos das linhas: ', countriesIdxType)
```
