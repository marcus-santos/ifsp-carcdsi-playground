```python
# Byscas no Dataframe

# Cria o DataFrame

datas = {'nome': ['Argentina', 'Brasil', 'França', 'Itália', 'Reino Unido'],
          'continente': ['América', 'América', 'Europa','Europa', 'Europa'],
          'extensao': [2780,8511,644,301,244],
          'corVerde': [0,1,0,1,0]
         }

codes = ['AR', 'BR', 'FR', 'IT', 'UK']

countries = pandas.DataFrame(datas, index=codes)

# Testa se um dado rótulo de linha existe

has_BR = 'BR' in countries.index
has_US = 'US' in countries.index
print("Existe o rótulo 'BR'? =>" , has_BR)
print("Existe o rótulo 'US'? =>" , has_US)
print("----------------------------------")

# Testa se um dado rótulo de coluna existe

has_greenColor = 'corVerde' in countries.columns
has_blueColor = 'corAzul' in countries.columns
print("Existe o rótulo 'corVerde'? => ", has_greenColor)
print("Existe o rótulo 'corAzul'? => ", has_blueColor)
print("----------------------------------")

# Testa se o valor faz parte de uma coluna

has_Brasil = countries['nome'].isin(['Brasil'])

print("Existe o valor 'Brasil' na coluna 'nome'? =>\n",has_Brasil)
```

```python
# Modificando o Dataframe

print('--- Dataframe original ---\n', countries)

# Inserindo o país Japão

countries.loc['JP'] = {
    'nome': 'Japão',
    'continente': 'Ásia',
    'extensao': '372',
    'corVerde': 0
    }

print('\n--- Dataframe com o Japão ---\n', countries)

# Altera a extensão do Brasil

countries.at['BR', 'extensao'] = 8512
print('\n---- Dataframe com extenão do Brasil modificada ---\n', countries)

# Remove a Argentina e o Reino Unido

countries = countries.drop(['AR', 'UK'])
print('\n --- Dataframe sem Argentina e Reino Unido ---\n', countries)
```
