```python
import pickle
import numpy as np
import pandas as pd
from sklearn.model_selection import GridSearchCV, cross_val_score, KFold
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
```

```python
from google.colab import drive
drive.mount('/content/drive')
caminho = '/content/drive/MyDrive/IFSP/Ciencia_Dados/Material'
```

```python
with open(caminho + '/credit.pkl', 'rb') as f:
    X_credit_treino, y_credit_treino, X_credit_teste, y_credit_teste = pickle.load(f)
```

```python
X_credit_teste.shape, y_credit_teste.shape
```

```python
X_credit_treino.shape, y_credit_treino.shape
```

```python
X_credit = np.concatenate((X_credit_treino, X_credit_teste), axis=0)
X_credit.shape
```

```python
y_credit = np.concatenate((y_credit_treino, y_credit_teste), axis=0)
y_credit.shape
```

### Árvore de Decisão

```python
parametros = {'criterion': ['gini', 'entropy'],
              'splitter': ['best', 'random'],
              'min_samples_split': [2, 5, 10],
              'min_samples_leaf': [1, 5, 10]}
```

```python
gri_search = GridSearchCV(estimator=DecisionTreeClassifier(), param_grid=parametros)
gri_search.fit(X_credit, y_credit)
melhores_parametros = gri_search.best_params_
melhor_resultado = gri_search.best_score_
print(melhores_parametros)
print(melhor_resultado)
```

### Random Forest

```python
parametros = {'criterion': ['gini', 'entropy'],
              'n_estimators': [10, 40, 100, 150],
              'min_samples_split': [2, 5, 10],
              'min_samples_leaf': [1, 5, 10]}
```

```python
gri_search = GridSearchCV(estimator=RandomForestClassifier(), param_grid=parametros)
gri_search.fit(X_credit, y_credit)
melhores_parametros = gri_search.best_params_
melhor_resultado = gri_search.best_score_
print(melhores_parametros)
print(melhor_resultado)
```

### KNN

```python
parametros = {'n_neighbors': [3, 5, 10, 20],
              'p': [1, 2]}
```

```python
gri_search = GridSearchCV(estimator=KNeighborsClassifier(), param_grid=parametros)
gri_search.fit(X_credit, y_credit)
melhores_parametros = gri_search.best_params_
melhor_resultado = gri_search.best_score_
print(melhores_parametros)
print(melhor_resultado)
```

### Regressão Logística

```python
parametros = {'tol': [0.0001, 0.00001, 0.000001],
              'C': [1.0, 1.5, 2.0],
              'solver': ['lbfgs', 'sag', 'saga']}
```

```python
gri_search = GridSearchCV(estimator=LogisticRegression(), param_grid=parametros)
gri_search.fit(X_credit, y_credit)
melhores_parametros = gri_search.best_params_
melhor_resultado = gri_search.best_score_
print(melhores_parametros)
print(melhor_resultado)
```

### SVM

```python
parametros = {'tol': [0.001, 0.0001, 0.00001],
              'C': [1.0, 1.5, 2.0],
              'kernel': ['rbf', 'linear', 'poly', 'sigmod']}
```

```python
gri_search = GridSearchCV(estimator=SVC(), param_grid=parametros)
gri_search.fit(X_credit, y_credit)
melhores_parametros = gri_search.best_params_
melhor_resultado = gri_search.best_score_
print(melhores_parametros)
print(melhor_resultado)
```

### Redes Neurais

```python
parametros = {'activation': ['relu', 'logistic', 'tanh'],
              'solver': ['adam', 'sgd'],
              'batch_size': [10,56]}
```

```python
gri_search = GridSearchCV(estimator=MLPClassifier(), param_grid=parametros)
gri_search.fit(X_credit, y_credit)
melhores_parametros = gri_search.best_params_
melhor_resultado = gri_search.best_score_
print(melhores_parametros)
print(melhor_resultado)
```

### Validação Cruzada

```python
resultados_arvore = []
resultados_random_forest = []
resultados_knn = []
resultados_regressao_logistica = []
resultados_svm = []
resultados_redes_neurais = []

for i in range(30):
  print(i)
  kfold = KFold(n_splits=10, shuffle=True, random_state=i)

  # {'criterion': 'entropy', 'min_samples_leaf': 1, 'min_samples_split': 5, 'splitter': 'best'}

  arvore = DecisionTreeClassifier(
      criterion='entropy',
      min_samples_leaf=1,
      min_samples_split=10,
      splitter='best')

  scores = cross_val_score(arvore, X_credit, y_credit, cv=kfold)
  resultados_arvore.append(scores.mean())

  # {'criterion': 'gini', 'min_samples_leaf': 1, 'min_samples_split': 2, 'n_estimators': 150}

  random_forest = RandomForestClassifier(
      criterion='gini',
      min_samples_leaf=1,
      min_samples_split=2,
      n_estimators=100)

  scores = cross_val_score(random_forest, X_credit, y_credit, cv=kfold)
  resultados_random_forest.append(scores.mean())

  # {'n_neighbors': 20, 'p': 1}

  knn = KNeighborsClassifier(n_neighbors=20, p=1)

  scores = cross_val_score(knn, X_credit, y_credit, cv=kfold)
  resultados_knn.append(scores.mean())

  # {'C': 1.0, 'solver': 'lbfgs', 'tol': 1e-05}

  regressao_logistica = LogisticRegression(C=1.0, solver='lbfgs', tol=0.00001)

  scores = cross_val_score(regressao_logistica, X_credit, y_credit, cv=kfold)
  resultados_regressao_logistica.append(scores.mean())

  # {'C': 2.0, 'kernel': 'rbf', 'tol': 0.001}

  svm = SVC(C=2.0, kernel='rbf', tol=0.001)

  scores = cross_val_score(svm, X_credit, y_credit, cv=kfold)
  resultados_svm.append(scores.mean())

  # {'activation': 'relu', 'batch_size': 10, 'solver': 'adam'}

  redes_neurais = MLPClassifier(activation='relu', batch_size=56, solver='adam')

  scores = cross_val_score(redes_neurais, X_credit, y_credit, cv=kfold)
  resultados_redes_neurais.append(scores.mean())
```

```python
resultados = pd.DataFrame({
    'Arvore': resultados_arvore,
    'Random Forest': resultados_random_forest,
    'KNN': resultados_knn,
    'Logistic Regression': resultados_regressao_logistica,
    'SVM': resultados_svm,
    'Redes Neurais: ': resultados_redes_neurais
})

resultados
```
